```python
object_feature = [col for col in data.columns if data[col].dtypes == 'object' ]
```

```python
df=pd.get_dummies(data, prefix=object_feature)
```

```python
data.info()
#X=data.iloc[:,0:10]
X=df.iloc[:,0:63]
X
```

```python
labelX=preprocessing.LabelEncoder()
y=labelX.fit_transform(y)
y
```

```python
#standardscale function
def standardScale(df):
    #run standardscaling
    stand=preprocessing.StandardScaler()
    df2=stand.fit_transform(df)
    #return array after standardscaling
    return df2
```

```python
X1=standardScale(X)
```

```python
# Feature Importance
from sklearn import datasets
from sklearn import metrics
from sklearn.ensemble import ExtraTreesClassifier
# load the iris datasets
# fit an Extra Trees model to the data
model = ExtraTreesClassifier()
model.fit(X1, y)
# display the relative importance of each attribute
print(model.feature_importances_)

```

```python
feature_list = pd.concat([pd.Series(X.columns), pd.Series(model.feature_importances_)], axis=1)
feature_list.columns = ['features_name', 'importance']
feature_list.sort_values("importance", ascending =False).head(20)
```
